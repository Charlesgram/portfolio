```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X=df.drop('sales',axis=1) #independent variables
y=df.sales #dependent variables
```

```python
 X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=101) #split into training and testing data
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model=LinearRegression() # calling an instance of the model
```

```python
model.fit(X_train,y_train) #fitting the training data into the model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
##predicting with the model
```

```python
y_pred = model.predict(X_test)
```

```python
##evaluating the model
```

```python
MAE = mean_absolute_error(y_test,y_pred)
```

```python
MAE
```

```python
R2 = r2_score(y_test,y_pred)
```

```python
resid = y_test-y_pred
```
